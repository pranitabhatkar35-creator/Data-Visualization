**Different operatons perform on array**

```python
import numpy as np
a=np.array([1,2,3,4])
b=np.array([1,2,3,4])
print(a-b)
print(a*b)
print(a//b)
```

**Broadcasting**

```python
a=np.array([1,2,3,4])
print(a+5)
print(a-1)
print(a/2)
```
