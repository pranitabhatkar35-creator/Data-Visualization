```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    "Title":["Extraction","The Crown","House of card","Bird Box","Stranger things","Enola Holmes","Queens","The lion king","Avtar","Bridgerton"],
    "Type":["Movie","TV Show","TV Show","Movie","TV Show","Movie","TV Show","Movie","TV Show","Movie"],
    "Genre":["Action","Drama","Thriller","Drama","Crime","Romance","Thrillear","Mystery","Mystery","Romance"],
    "Release_year":[2020,2016,2021,2019,2020,2022,2023,2020,2024,2025],
}
df=pd.DataFrame(data)

plt.figure(figsize=(6,3))
df["Type"].value_counts().plot(kind="bar",color=["SKYBLUE","PINK",],edgecolor="black")
plt.title("Movieas vs Tv show")
plt.ylabel("Count")
plt.tight_layout(pad=3)
plt.grid(axis="y",linestyle="--",alpha=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    "Title":["Extraction","The Crown","House of card","Bird Box","Stranger things","Enola Holmes","Queens","The lion king","Avtar","Bridgerton"],
    "Type":["Movie","TV Show","TV Show","Movie","TV Show","Movie","TV Show","Movie","TV Show","Movie"],
    "Genre":["Action","Drama","Thriller","Drama","Crime","Romance","Thrillear","Mystery","Mystery","Romance"],
    "Release_year":[2020,2016,2021,2019,2020,2022,2023,2020,2024,2025],
}
df=pd.DataFrame(data)

plt.figure(figsize=(5,3))
df["Genre"].value_counts().plot(kind="bar",color="skyblue",edgecolor="black")
plt.title("Top Genres on Netflix")
plt.ylabel("count")
plt.xticks(rotation=45,ha="right")
plt.tight_layout(pad=3)
plt.grid(axis="y",linestyle="--",alpha=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    "Title":["Extraction","The Crown","House of card","Bird Box","Stranger things","Enola Holmes","Queens","The lion king","Avtar","Bridgerton"],
    "Type":["Movie","TV Show","TV Show","Movie","TV Show","Movie","TV Show","Movie","TV Show","Movie"],
    "Genre":["Action","Drama","Thriller","Drama","Crime","Romance","Thrillear","Mystery","Mystery","Romance"],
    "Release_year":[2020,2016,2021,2019,2020,2022,2023,2020,2024,2025],
}
df=pd.DataFrame(data)

plt.figure(figsize=(5,3))
df["Release_year"].value_counts().sort_index().plot(marker="o",color="green")
plt.title("Release trend over years")
plt.xlabel("Release Year")
plt.ylabel("Number of titles")
plt.grid(True,linestyle="--",alpha=0.5)
plt.tight_layout(pad=3)
plt.show()
```
