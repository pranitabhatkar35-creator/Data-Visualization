```python
import matplotlib.pyplot as plt
x=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
y=[7,10,15,20,12,25]
plt.plot(x,y)
plt.tittle("Bakery Sales")
plt.xlable("Day of the week")
plt.ylabel("sales per day")
plt.show()
```

```python
#Bar Chart
import matplotlib.pyplot as plt
Product=["Milk","Curd","Paneer","Butter"]
Sales=[1000,1200,1400,900]
plt.bar(Product,Sales,color="Green",label="sale2025")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Sales Reports")
plt.legend()
plt.show()
```

```python
#Pie chart
plt.pie(Sales,labels=Product,autopct="%1.1f%%")
plt.show()
```

```python
import matplotlib.pyplot as plt
Scores=[45,67,89,78,88,92,60,54,78,79,82,84,88,66,85]
plt.hist(Scores,bins=5,color="Purple",edgecolor="Black")
plt.xlabel(Scores)
plt.ylabel("Number of Students")
plt.title("Test Scores")
plt.show()

```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,2,figsize=(10,5))
x=[1,2,3,4]
y=[10,20,15,25]
ax[0].plot(x,y,color="blue")
ax[0].set_title("Line Chart")
ax[1].bar(x,y,color="green")
ax[1].set_title("Bar Chart")
fig.suptitle("Subplots Example")
plt.tight_layout()
plt.show()
```
