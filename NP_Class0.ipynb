```python
import numpy as np
```

```python
a=[1,2,3,4,5]
print(a)
arr=np.array(a)
print("This is Array :",arr)
```

```python
my_list=[[1,2,3],[4,5,6],[7,8,9]]
print(f"This is a list:{my_list}its type is {type(my_list)}")
my_arrMatrix=np.array(my_list)
print(f"This is matrix : \n{my_arrMatrix}\n its type is {my_arrMatrix.dtype}")
```

```python
vector=np.zeros(3)
print(f"This is a vector \n{vector}")
matrix=np.zeros((3,3))
print(f"This is a matrix \n{matrix}")
```

```python
vector=np.ones(3)
print(f"This is a vector \n{vector}")
matrix=np.ones((3,3))
print(f"This is a matrix \n{matrix}")
```

```python
vector=np.full(3,5)
print(f"This is a vector \n{vector}")
matrix=np.full((3,3),5)
print(f"This is a matrix \n{matrix}")
```
